## Load data from Hugging Face (risks dataset Sovai)

### insert data from Hugging Face

```python
import pandas as pd 
import requests
url = "https://datasets-server.huggingface.co/rows?dataset=sovai%2Frisk_indicators&config=default&split=train&offset=0&length=100"
ss=requests.get(url)
print(ss.status_code)
```

### what is inside dataset

```python
data=ss.json()
data.keys()
```

```python
data['num_rows_total']
```

```python
data['features']
```

```python
data[ 'rows']
```

### Make it tabular:

```python

column_names = [feature['name'] for feature in data['features']]
rows = [r['row'] for r in data['rows']]
df=pd.DataFrame(rows , columns=column_names)
df

```

```python
df.to_csv('risk_indicators.csv')
```
